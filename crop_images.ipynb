```
import numpy as np # linear algebra
import xml.etree.ElementTree as ET # for parsing XML
import matplotlib.pyplot as plt # to show images
from PIL import Image # to read images
import os
import glob
```

```
root_images="dataset/flower/train_imgs/"
root_annots="dataset/flower/train_anns/"
root_save="dataset/flower/"
directory="flower_cropped"
```

```
all_images=os.listdir("dataset/flower/train_imgs/")
print(f"Total images : {len(all_images)}")

breeds = glob.glob('dataset/flower/train_anns/')
annotation=[]
for b in breeds:
    annotation+=glob.glob(b+"/*")
print(f"Total annotation : {len(annotation)}")

breed_map={}
for annot in annotation:
    breed=annot.split("/")[-2]
    index=breed.split("-")[0]
    breed_map.setdefault(index,breed)
    
print(f"Total Breeds : {len(breed_map)}")
print(breed_map)
```

```
def bounding_box(image):
    bpath=root_annots+str(image.split(".")[0])+".xml"
    tree = ET.parse(bpath)
    root = tree.getroot()
    objects = root.findall('object')
    print(len(objects))
    item = len(objects)
    for o in objects:
        bndbox = o.find('bndbox') # reading bound box
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

        save_name = str(image.split(".")[0]) + '-' + str(item)
        # print(save_name, xmin, ymin, xmax, ymax)
        im=Image.open(os.path.join(root_images,image))
        # bbox = xmin, ymin, xmax, ymax
        im=im.crop((xmin, ymin, xmax, ymax))
        # save_file = open(os.path.join(root_save, directory, save_name), 'w')
        im.save(os.path.join(root_save, directory, save_name), "JPEG")

        item = item - 1
        # save_file.close()
        # plt.subplot(5,5,i+1)
        # plt.figure(figsize=(10,10))
        # plt.subplot(5,5,i+1)
        # plt.axis("off")
        # plt.imshow(im) 
    
    return (len(objects))
```

```
plt.figure(figsize=(10,10))
total = 0
for i,image in enumerate(all_images):
    #bbox=bounding_box(image)
    # im=Image.open(os.path.join(root_images,image))
    # im=im.crop(bbox)
    
    # plt.subplot(5,5,i+1)
    # plt.axis("off")
    # plt.imshow(im)    
    # if(i==1):
    #     break

    crop = bounding_box(image)
    total = total + crop
    print(f"Total Annotations : ", total)
```
